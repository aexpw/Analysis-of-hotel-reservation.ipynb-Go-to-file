```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv('/Users/god/Desktop/bookings.csv', sep=';') 
```

```python
# Функция заменяет пробел на нижнее подчеркивание, делает шрифт строчным
def to_lower_underscore(name):
    name = name.lower().replace(' ','_')
    return name 
df = df.rename(columns = to_lower_underscore)
```

```python
df
```

```python
# Сравнение среднего кол-во успешно забронированных ночей в зависимости от отеля 
df.groupby('hotel').agg({'stays_total_nights': 'mean'}).round(4)
```

```python
# Кол-во отмененных бронирований в отелях City Hotel по годам

df_years = df[(df['hotel'] == 'CityHotel')  & (df['is_canceled'] == 1)]\
            .groupby('arrival_date_year', as_index=False) \
            .agg({'is_canceled' : 'sum'}) 
df_years
```

```python
# Кол-во отмененных бронирований в какждом месяце каждого года

df_years_1 = (df.query('is_canceled == 1')
.groupby(['arrival_date_year', 'arrival_date_month'], as_index = False) 
.agg({'is_canceled' : 'sum'}) 
.sort_values('is_canceled', ascending = False))


df_years_1
            
```

```python
# Средний поакзатель наличия взрослых, детей, маленьких детей, во всех бронированиях
df[['adults','children', 'babies']].mean()

```

```python
# Добавление в исходую таблицу столбца total_kids с указанием общего количества детей в каждом бронировании
df['total_kids'] = df['children'] + df['babies']
df[df['total_kids'] > 0 ]
```

```python
# Среднее кол-во бронирований с детьми в каждом типе отеля
df.query('total_kids > 0') .groupby('hotel', as_index = False) .agg({'total_kids' : 'mean'})
```

```python
# Кол-во не отмененных и отмененных бронирований в каждой возрастной группе детей
pd.crosstab(df['is_canceled'], df['total_kids'])
```

```python
# Добавление в таблицу столбца has_kids со значениями True или False
df['has_kids'] = df['total_kids'].apply(lambda x: True if x else False)
```

```python
# бщее кол-во бронирований с детьми 
number_of_childrenn = 0
for i in df['total_kids']:
    if  i == True :
        number_of_childrenn += 1  
print(number_of_childrenn )
```

Проведем расчет churn rate (коэффициента оттока клиентов) 
Посмотрим как влияет отмена бронирования в зависимости от наличия или отсутсвия детей

```python
churn_without = round((df.query("has_kids == False & is_canceled == 1").shape[0]
       / df.query("has_kids == False").shape[0])*100,2) # Отток клиентов без детей

churn_with = round((df.query("has_kids == True & is_canceled == 1").shape[0]
       / df.query("has_kids == True").shape[0])*100,2) # Отток клиентов с детьми
```

```python
churn_without,\
churn_with
```

```python
round(churn_without - churn_with, 2 )
```

Отток клиентов с детьми меньше , чем отток клиентов без детьми.
Разница составляет 2,27 %
